```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
df = load_iris() #Load dataset 
data = pd.DataFrame(df.data, columns = df.feature_names)
data["target"] = df.target
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[df.feature_names], data["target"], random_state = 0)
```

```python
clf = DecisionTreeClassifier(max_depth = 2, random_state = 0)
clf.fit(x_train, y_train) #perfroming the training
y_pred = clf.predict(x_test) #answers you get after prediction on the x_test
```

```python
tree.plot_tree(clf)
```

```python
fn = ["sepal length", "sepal width", "petal length", "petal width"]
cn = ["setosa", "versicolor", "virginica"]
fig, axis = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True)
fig.savefig("image.png")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
